```python
def fact():
  get=int(input("Enter the number"))
  if get==0:
    print("The factorial of the number is 1")
  elif get==1:
    print("The factorial of the number is 1")
  else:
    fact_result=1
    for i in range(1,get+1):
      fact_result *=i
    print("The factorial of the number is ",fact_result)
```

```python
fact()
```

```python
def even():
  get2=int(input("Enter the variable"))
  if get2==0:
    print("The number is 0")
  elif get2 % 2==0:
    print("The number is even")
  else:
    print("The number is odd")
```

```python
even()
```
