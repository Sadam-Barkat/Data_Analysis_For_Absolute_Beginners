```python
df_movies = pd.read_excel('/content/movies_db.xlsx', 'movies')
df_movies.head()
```

```python
def standard_currency(curr):
  if curr == "$$" or curr == "Dollars":
    return "USD"
  else:
    return curr

df_financials = pd.read_excel('/content/movies_db.xlsx', 'financials', converters = {
    'currency': standard_currency
})
df_financials.head()
```

```python
df_merge = pd.merge(df_movies, df_financials, on='movie_id')
df_merge.head()
```

```python
df_merge.to_excel('merge2.xlsx', sheet_name = 'new_movies', index=False)
```
