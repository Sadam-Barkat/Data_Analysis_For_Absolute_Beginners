#Understanding min() and max() with other example

```python
import pandas as pd

# Sample Data
data = {
    'day': ['1/1/2017', '1/2/2017', '1/3/2017', '1/4/2017',
            '1/1/2017', '1/2/2017', '1/3/2017', '1/4/2017',
            '1/1/2017', '1/2/2017', '1/3/2017', '1/4/2017'],
    'city': ['new york', 'new york', 'new york', 'new york',
             'mumbai', 'mumbai', 'mumbai', 'mumbai',
             'paris', 'paris', 'paris', 'paris'],
    'temperature': [32, 36, 28, 33, 90, 85, 87, 92, 45, 50, 54, 42],
    'windspeed': [6, 7, 12, 7, 5, 12, 15, 5, 20, 13, 8, 10],
    'event': ['Rain', 'Sunny', 'Snow', 'Sunny', 'Sunny', 'Fog', 'Fog', 'Rain', 'Sunny', 'Cloudy', 'Cloudy', 'Cloudy']
}

df = pd.DataFrame(data)

# Group by city and get the minimum values
group = df.groupby('city').min()

# Print the minimum value for a specific column (e.g., 'temperature')
print(group['day'])
```

```python
import pandas as pd

# Sample Data
data = {
    'day': ['1/1/2017', '1/2/2017', '1/3/2017', '1/4/2017',
            '1/1/2017', '1/2/2017', '1/3/2017', '1/4/2017',
            '1/1/2017', '1/2/2017', '1/3/2017', '1/4/2017'],
    'city': ['new york', 'new york', 'new york', 'new york',
             'mumbai', 'mumbai', 'mumbai', 'mumbai',
             'paris', 'paris', 'paris', 'paris'],
    'temperature': [32, 36, 28, 33, 90, 85, 87, 92, 45, 50, 54, 42],
    'windspeed': [6, 7, 12, 7, 5, 12, 15, 5, 20, 13, 8, 10],
    'event': ['Rain', 'Sunny', 'Snow', 'Sunny', 'Sunny', 'Fog', 'Fog', 'Rain', 'Sunny', 'Cloudy', 'Cloudy', 'Cloudy']
}

df1 = pd.DataFrame(data)

# Group by city and get the minimum values
group = df1.groupby('city').max()

# Print the minimum value for a specific column (e.g., 'temperature')
print(group['temperature'])
```

#Understanding min() and max() with other example

```python
import pandas as pd

# Create a DataFrame with only numerical data
data = {
    'Math': [85, 90, 78, 92],
    'Science': [88, 76, 95, 89],
    'English': [90, 85, 88, 92]
}

df = pd.DataFrame(data)

# Calculate the mean of each column
mean_columns = df.mean()
print("Mean of each column:")
print(mean_columns)

# Calculate the mean of each row (axis=1)
mean_rows = df.mean(axis=1)
print("\nMean of each row:")
print(mean_rows)

# Calculate the overall mean of the entire DataFrame
overall_mean = df.mean().mean()
print(f"\nOverall mean of the DataFrame: {overall_mean}")
```

#replace()

```python
import pandas as pd

# Sample dataset
data = {
    'Name': ['Alice', 'Bob', None, 'David', 'Eve'],
    'Age': [25, None, 30, 999, -1],
    'City': ['New York', 'NA', 'San Francisco', 'Unknown', 'Chicago']
}

df = pd.DataFrame(data)
print("Original Dataset:")
print(df)
```

```python
# Replace specific values
df_replaced = df.replace({'Age': 999}, {'Age': None})
print("\nDataset after replacing 999 with None in 'Age':")
print(df_replaced)
```

#replace multiple values

```python
# Replace multiple invalid values
df_cleaned = df.replace({
    'Age': {999: None, -1: None},
    'City': {'NA': None, 'Unknown': None}
})
print("\nDataset after replacing multiple invalid values:")
print(df_cleaned)
```

#Simple concat()

```python
import pandas as pd

india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

#Concatenation with correct indexing

```python
df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

#Simple merge()

```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3 = pd.merge(df1, df2, on="city")
df3
```
