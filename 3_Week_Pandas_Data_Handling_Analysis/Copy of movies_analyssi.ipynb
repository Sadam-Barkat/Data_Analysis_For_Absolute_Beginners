```python
import pandas as pd
```

```python
df_movies = pd.read_excel('/content/movies_db.xlsx', sheet_name = 'movies')
df_movies.head()
```

```python
df_financials = pd.read_excel('/content/movies_db.xlsx', sheet_name = 'financials')
df_financials.head()
```

```python
df_movies_df_financials = pd.merge(df_movies, df_financials, on= 'movie_id')
df_movies_df_financials.head()
```

```python
df_languages = pd.read_excel('/content/movies_db.xlsx', sheet_name = 'languages')
df_languages
```

```python
data = pd.merge(df_movies_df_financials, df_languages, on ='language_id')
data.head()
```

```python
data = data[['title', 'industry', 'release_year', 'imdb_rating', 'studio', 'name','budget', 'revenue', 'unit', 'currency']]
data.head()
```
