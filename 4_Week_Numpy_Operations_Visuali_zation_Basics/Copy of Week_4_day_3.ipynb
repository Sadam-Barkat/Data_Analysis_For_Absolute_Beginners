#Introduction to Seaborn
**What is Seaborn?**
A Python library built on top of Matplotlib.
Provides a high-level interface for creating attractive and informative visualizations.

**Why use Seaborn?**
Simplifies visualization creation.
Integrates well with Pandas DataFrames.
Includes themes and color palettes.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load a built-in dataset
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='time')
plt.title('Scatter Plot: Total Bill vs Tip')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='time', markers={'Lunch': 's', 'Dinner': 's'})
plt.title('Scatter Plot: Total Bill vs Tip')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
plt.show()
```

```python
sns.lineplot(data=tips, x='size', y='tip')
plt.title('Line Plot: Size vs Tip')
plt.show()
```

```python
sns.lineplot(data=tips, x='size', y='tip', errorbar=None)
plt.title('Line Plot: Size vs Tip')
plt.show()
```

```python
sns.histplot(data=tips, x='total_bill', bins=20, kde=True)
plt.title('Histogram of Total Bills')
plt.show()
```

```python
sns.barplot(data=tips, x='day', y='total_bill', hue='sex')
plt.title('Average Total Bill by Day and Gender')
plt.show()
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
plt.title('Box Plot: Total Bill by Day and Gender')
plt.show()
```

```python
sns.set_style('whitegrid')  # Options: darkgrid, whitegrid, dark, white, ticks
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', palette='Set2', )
plt.show()
```

```python
plt.title('Box Plot: Total Bill')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex')
plt.show()
```

```python
# Create a correlation matrix, only including numerical features
numerical_features = tips.select_dtypes(include=np.number).columns
corr = tips[numerical_features].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlations')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset('tips')

# Scatter plot
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')

# Add a title
plt.title('Scatter Plot: Total Bill vs Tip', fontsize=14, fontweight='bold')
plt.show()
```

```python
sns.barplot(data=tips, x='day', y='total_bill', hue='sex')

# Add axis labels
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Total Bill (in USD)', fontsize=12)

# Add a title
plt.title('Average Total Bill by Day and Gender', fontsize=14, fontweight='bold')

plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time')

# Customize the legend
plt.legend(title='Legend', loc='upper left', fontsize=10, title_fontsize=12)

# Add a title
plt.title('Scatter Plot: Total Bill vs Tip (by Gender and Time)', fontsize=14)

plt.show()
```
