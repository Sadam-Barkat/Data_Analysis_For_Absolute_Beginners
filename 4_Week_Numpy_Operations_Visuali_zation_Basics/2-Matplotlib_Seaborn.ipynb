```python
import matplotlib.pyplot as plt
x=['Apple','Samsung','Xiaomi','Huawei','oppo']
y=[150,200,180,170,160]
c=['purple','blue','green','brown','red']

plt.xlabel("smartphone brands",fontsize=14)
plt.ylabel("sales(in million)",fontsize=14)
#plt.title("smartphone sales comparisons")
plt.bar(x,y,width=0.3,color=c,align='edge',edgecolor='black',linestyle=':',linewidth=2,alpha=0.5,label="popularity")
plt.legend()
plt.show()
```

```python
import numpy as np
x=['Apple','Samsung','Xiaomi','Huawei','oppo']
y=[150,200,180,170,160]
y2=[70,120,140,100,50]
width=0.2
gap=0.3
p=np.arange(len(x))
p1=[i+width+gap for i in p]

plt.xlabel("smartphone brands",fontsize=14)
plt.ylabel("sales(in million)",fontsize=14)
#plt.title("smartphone sales comparisons")
plt.bar(p,y,color='red',width=width,align='edge',edgecolor='black',linewidth=2,alpha=0.5,label="2021")
plt.bar(p1,y2,color='green',width=width,align='edge',edgecolor='black',linewidth=2,alpha=0.5,label="2022")
plt.legend()
plt.xticks(p+width,x)
plt.show()
```

Scatter **Graph**

```python
prices=[10000,20000,300000,500000]
rating=[4.5,5,3,4]
c=['red','yellow','green','brown']
size=[30,40,50,60]
plt.xlabel("Mobile Price ",fontsize=20)
plt.ylabel("rating")
#plt.title("price vs rating")
plt.scatter(prices,rating,color=c,s=size,marker='+')
plt.show()
```
