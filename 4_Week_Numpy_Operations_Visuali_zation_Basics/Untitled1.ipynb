```python
import plotly.express as px
```

```python
data = px.data.tips()
data.head()
```

# **BarPlot**

```python
fig = px.bar(data, y= 'tip', x='time', color='sex')
fig.show()
```

```python
fig = px.bar(data, x='day', y='total_bill', color='smoker')
fig.show()
```

```python
query_data = data.query("total_bill > 15.00")
px.bar(query_data, x='sex', y='total_bill')
```

```python
query_data = data.query("total_bill > 15.00 and day== 'Sun'")
px.bar(query_data, x='day', y='total_bill', color = 'sex')
```

```python
df_condition = data[data.time=='Dinner']
px.bar(df_condition, x='sex', y='total_bill', hover_data = ['tip','day'], color='smoker').update_traces(width=0.2).update_layout(title="bargraph", title_font_weight = 'bold').show()
```

# **Scatter**

```python
fig = px.scatter(data, x='total_bill', y='tip', color='day', size='size', hover_data = ['sex', 'time'])
fig.update_layout(
    title='graph',
    title_x=0.5,
    title_font_size = 20,
    title_font_family = 'Arial',
    title_font_weight = 'bold',
    xaxis_title = 'one',
    yaxis_title='two').show()
```

# **PieChart**

```python
fig = px.pie(data, values='total_bill', names='day', color='smoker')
fig.show()
```

```python
fig = px.pie(data, values='total_bill', names='day', color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```
