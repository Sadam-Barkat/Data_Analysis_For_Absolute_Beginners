## Dict Comprehension

```python
squares = {x:x*x for x in range(1,6)}
print(squares)
```

```python
num = {1:2, 3:4, 5:6}
num
```

```python
sales_data = [
    {"product":"A", "sale":150},
    {"product":"B", "sale":400},
    {"product":"C", "sale":300},
    {"product":"A", "sale":250},
    {"product":"B", "sale":50}
]

total_sales = {}

for sale in sales_data:
  product = sale['product']
  sale_amount = sale['sale']

  if product in total_sales:
    total_sales[product] += sale_amount
  else:
    total_sales[product] = sale_amount
```

```python
total_sales
```
